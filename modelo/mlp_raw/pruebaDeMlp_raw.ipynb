```python
from tensorflow.keras.models import load_model
import pandas as pd

# Cargar modelo
modelo = load_model("modelo_poker_raw.keras")

# Nueva mano
nueva_mano = pd.DataFrame([{
    'S1':1,'C1':10,
    'S2':1,'C2':11,
    'S3':1,'C3':12,
    'S4':1,'C4':13,
    'S5':1,'C5':1
}])

# Convertir a array (sin normalizar)
# X_nueva = nueva_mano.drop(columns=['CLASS']).values
X_nueva = nueva_mano.values

# Predecir
pred = modelo.predict(X_nueva)
clase = pred.argmax(axis=1)



significado = {
    0: "Nothing in hand",
    1: "One pair",
    2: "Two pairs",
    3: "Three of a kind",
    4: "Straight",
    5: "Flush",
    6: "Full house",
    7: "Four of a kind",
    8: "Straight flush",
    9: "Royal flush"
}


print("Clase predicha:", significado[clase[0]])
```
