```python
import sys
import os
import joblib

# Ajustar path correctamente (subir hasta raíz del proyecto)
sys.path.append(os.path.abspath("../.."))

from tensorflow.keras.models import load_model
from preprocesamiento.normalizacionDelDataset import normalizarDatasetValidationTest
import pandas as pd
```

```python
modelo = load_model("modelo_poker.keras")

scaler = joblib.load("../../modelo/baseline/scaler.pkl")
encoder = joblib.load("../../modelo/baseline/encoder.pkl")

# Nueva mano
nueva_mano = pd.DataFrame([{
    'S1':1,'C1':13,
    'S2':2,'C2':13,
    'S3':3,'C3':13,
    'S4':4,'C4':5,
    'S5':1,'C5':9
}])

# Preprocesar
nueva_mano_procesada = normalizarDatasetValidationTest(
    nueva_mano.assign(CLASS=0),
    scaler,
    encoder
)

X_nueva = nueva_mano_procesada.drop(columns=['CLASS']).values

# Predecir
pred = modelo.predict(X_nueva)
clase = pred.argmax(axis=1)

print("Clase predicha:", clase)
```
